## 지도 시각화
- 지도 생성하기
    - m = folium.Map(location = [위도, 경도], zoom_start = 확대정도)
- 정보 추가하기
    - 마커 추가하기
        - folium.Marker([위도, 경도]).add_to(m)
    -  원 추가하기
        - folium.CircleMarker([위도, 경도], radious= 원크기).add_to(m)
    -  추가옵션:
        - tooltip="마우스 올리면 보여질 정보"
        - popup="클릭하면 보여질 정보"

#### 지도 생성하기
지도를 만들어야 하는데, 위도와 경도값을 기준으로 수행한다. 구글 지도를 활용해서 위도와 경도값을 얻어오도록 한다.<br>연세대학교: [37.56592004772955, 126.93858272691241]<br>
zoom_start 옵션을 통해, 얼마나 확대해서 볼 것인지를 정한다.

```python
m = folium.Map(location = [37.56592004772955, 126.93858272691241], zoom_start=10)
m
```

```python
m = folium.Map(location = [37.56592004772955, 126.93858272691241], zoom_start=15)
m
```

#### 마커 추가하기
연세대학교에 마커를 추가하도록 한다.

```python
folium.Marker([37.56592004772955, 126.93858272691241]).add_to(m)
m
```

마커에 핀만 표시되는 것이 아닌 추가정보를 부여할 수 있다.

```python
m = folium.Map(location = [37.56592004772955, 126.93858272691241], zoom_start=15)

folium.Marker([37.56592004772955, 126.93858272691241],
             tooltip='마우스를 올릴 때 보여지는 정보').add_to(m)
m
```

```python
m = folium.Map(location = [37.56592004772955, 126.93858272691241], zoom_start=15)

folium.Marker([37.56592004772955, 126.93858272691241],
             tooltip='마우스를 올릴 때 보여지는 정보',
             popup='클릭할 때 보여지는 정보').add_to(m)
m
```

#### 써클마커 추가하기

일반마커와는 다르게, 반드시 radius를 추가해야 한다.

```python
folium.CircleMarker([37.56592004772955, 126.93858272691241],
                   radius=20).add_to(m)
m
```

### 미니맵을 추가할 경우 MiniMap

folium에는 다양한 효과를 줄 수 있는 plugin이 존재하는데, Minimap은 그 중 하나다.
미니맵을 지도 우측 하단에 추가하여, 현재 어느 위치를 살펴보고 있는지 점검할 수 있다.

```python
from folium.plugins import MiniMap

m = folium.Map(location = [37.56592004772955, 126.93858272691241], zoom_start=12)

#미니맵추가
minimap = MiniMap()
minimap.add_to(m)

m
```

### 서울 대피소 현황 지도 만들기

```python
import pandas as pd
```

서울열린데이터광장에 있는 서울시 대피소 현황 자료를 사용해서 지도를 만들어봅니다.

```python
file = './data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'

raw = pd.read_csv(file, encoding='cp949')
raw.head()
```

```python
  raw.info()
```

```python
#인덱스번호가 0인 경우의 위도와 경도 데이터를 출력해본다.
#loc: 인덱스, column
lat, lon = raw.loc[0, ['위도','경도']]
print(lat,lon)
```

```python
#특정 인덱스번호에 대해 위도, 경도, 명칭을 출력한다.
import random

i = random.randint(0, raw.shape[0])

lat, lon, name = raw.loc[i,['위도', '경도','대피소명칭']]
print(lat, lon, name)
```

지도를 생성하고 위도/경도에 맞게 지도에 마커를 추가한다.

```python
#지도 생성하기
m = folium.Map(location= [37.56592004772955, 126.93858272691241], zoom_start=12)

#마커 생성하기
for i in range(raw.shape[0]):
    lat, lon, name = raw.loc[i,['위도', '경도','대피소명칭']]
    folium.Marker([lat,lon]).add_to(m)

m
```

마커에 마우스를 올리면 이름이 뜨도록 해봅시다.

```python
#지도 생성하기
m = folium.Map(location= [37.56592004772955, 126.93858272691241], zoom_start=12)

#마커 생성하기
for i in range(raw.shape[0]):
    lat, lon, name = raw.loc[i,['위도', '경도','대피소명칭']]
    folium.Marker([lat,lon], tooltip=name).add_to(m)

m
```

지도를 저장하는 방법에 대해 알아봅시다.

```python
m.save('./data/Sheltermap.html')
```

### ClusterMarker

마커가 너무 많은 경우에는 보기 힘들기 때문에, ClusterMarker를 사용하게 됩니다.
근처에 있는 마커를 그룹화해서 봅시다.

```python
from folium.plugins import MarkerCluster

#지도 생성하기
m = folium.Map(location= [37.56592004772955, 126.93858272691241], zoom_start=12)

marker_cluster = MarkerCluster().add_to(m)

#마커 생성하기
for i in range(raw.shape[0]):
    lat, lon, name = raw.loc[i,['위도', '경도','대피소명칭']]
    folium.Marker([lat,lon], tooltip=name).add_to(marker_cluster)

# 줌인, 줌아웃을 하거나, 클러스터를 선택해 자세히 볼 수 있다.
m
```
