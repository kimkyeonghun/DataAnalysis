## 자료 받아오기(Open API 활용)

```python
#API를 사용하기 위한 라이브러리
import requests
import pandas as pd
```

```python
#apikey 입력하기
apikey = '/your/api/key'
```

요청할 URL 주소를 만든다.
<br>한번에 최대 1천개까지만 가능하기 때문에 첫 1000개의 데이터를 가져옵니다.

```python
url = f'http://openapi.seoul.go.kr:8088/{apikey}/json/bikeList/1/1000/'
url2 = f'http://openapi.seoul.go.kr:8088/{apikey}/json/bikeList/1001/2000/'
```

```python
#제대로 연결된 것을 알 수 있음
requests.get(url)
```

```python
json1 = requests.get(url).json()
json2 = requests.get(url2).json()
```

필요한 정보를 선택한다.

```python
json1.keys()
```

```python
json1['rentBikeStatus']
```
