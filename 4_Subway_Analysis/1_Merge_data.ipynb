### 폴더의 모든 파일 병합하기
위처럼 파일을 하나씩 합치는 것은 비효율적이기 때문에 파이썬의 기능을 사용해 모두 병합한다.

```python
import os
```

```python
os.listdir('./rawfiles')
```

```python
raw = pd.DataFrame()
file_dir = './rawfiles'

for i, file in enumerate(os.listdir(file_dir)):
    temp = pd.read_csv(os.path.join(file_dir,file))
    raw = raw.append(temp)
    print(f"{i} th append: ", raw.shape[0])
    
raw
```

```python
raw.info()
```

인덱스가 각 파일의 인덱스 값으로 설정되어 있기 때문에 새로 설정하도록 한다.

```python
raw = raw.reset_index(drop=True)
raw
```

### 요일데이터 추가하기

```python
from datetime import datetime
```

datetime.strptime('날짜', `str형태`): 문자 --> 날짜 타입
- `str형태`: %Y-%m-%d %H:%M:%S
    - %Y: 연도(4자리)
    - %m: 월(2자리)
    - %d: 일자(2자리)
    - %H: 시간
    - %M: 분
    - %S: 초

```python
weekday_dict = ['월','화','수','목','금','토','일']
weekday = []
for date in raw['사용일자']:
    d = datetime.strptime(str(date),'%Y%m%d')
    weekday_index = d.weekday()
    weekday.append(weekday_dict[weekday_index])
    
raw['요일'] = weekday
raw.head()
```
